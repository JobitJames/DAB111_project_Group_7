```python
import csv
import sqlite3
```

```python
con = sqlite3.connect('Canada_Climate_Dataset.db')
cursor = con.cursor()
```

```python
create_table = """CREATE TABLE IF NOT EXISTS ClimateStations (
                    x FLOAT,
                    y FLOAT,
                    STN_ID INT PRIMARY KEY,
                    STATION_NAME TEXT,
                    PROV_STATE_TERR_CODE TEXT,
                    ELEVATION FLOAT,
                    first_year INT,
                    last_year INT
                );"""

cursor.execute(create_table)
```

```python
with open("selected_stations.csv", "r") as f:
    data = csv.reader(f)
    header = next(data)  # Skip the header row
    for row in data:
        cursor.execute("""
        INSERT INTO ClimateStations (x, y, STN_ID, STATION_NAME, PROV_STATE_TERR_CODE, ELEVATION, first_year, last_year)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
        """, (
            float(row[0]), float(row[1]),int(row[2]),row[3],row[4],float(row[5]),int(row[6]),int(row[7])
        ))
```

```python
con.commit
```

```python
cursor.execute("SELECT * FROM ClimateStations").fetchall()
```
